```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(tips['total_bill'], kde=True, bins=30)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.pairplot(tips)
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
import numpy as np

sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips, palette='deep')
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.set_style('ticks')
sns.countplot(x='sex', data=tips, palette='deep', ax=ax)

ax.set_xlabel("Eixo x")
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex', palette='rainbow')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='sex', split=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['x', 'v'], scatter_kws={'s':100}, aspect=1.6)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='time', row='day')
```

```python
sns.lmplot(x='total_bill', y='tip', aspect=4, data=tips)
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
pivotFlights = flights.pivot_table(index='year', columns='month', values='passengers')
pivotFlights
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
sns.heatmap(pivotFlights, ax=ax, annot=True, fmt='.0f', linewidths=1, linecolor='white', cmap='magma')
```

```python
sns.clustermap(pivotFlights, standard_scale=0.5)
```
