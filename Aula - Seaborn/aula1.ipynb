```python
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(tips['total_bill'], kde=True, bins=30)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.pairplot(tips)
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
import numpy as np

sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex', palette='rainbow')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='sex', split=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['x', 'v'], scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='time')
```
